```python
# import necessary libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading the data as pandas DataFrame

movies=pd.read_csv("movies.csv")
credits=pd.read_csv("credits.csv")
```

```python
credits.head()
```

```python
# merging 'movies' and 'credits' DataFrame

movies=pd.merge(movies,credits,left_index=True,right_index=True)
movies.head()
movies.info()
```

```python
# fetch only relevant columns from the DataFrame
movies=movies[['id','title_x','overview','genres','keywords','cast','crew']]

# dropping rows with NULL values
movies.dropna(inplace=True)
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.iloc[0].genres
```

```python
# import ast library to convert string to object

import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L 
```

```python
# Converting strings to objects in 'genres' columns

movies.genres=movies.genres.apply(convert)
movies.head()


# Converting strings to objects in 'keywords' columns
movies.keywords=movies.keywords.apply(convert)
```

```python
# extracting important casts from casts column

def convert_cast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if(counter==3): break
        L.append(i['name'])
        counter+=1
    return L

movies.cast=movies.cast.apply(convert_cast)
movies.head()
```

```python
# extracting the name of director from the crew dictionary

def convert_crew(obj):
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            return i['name']
        
movies.crew=movies.crew.apply(convert_crew)
movies.head()
```

```python
movies['overview'][0]
```

```python
# splitting the string into list in overview column

movies['overview']=movies['overview'].apply(lambda x:str(x).split())
type(movies['overview'][0])
movies.head()
```

```python
#remove spaces from the cast and crew names -- to easily create tags

movies.dropna(inplace=True)
movies['keywords']=movies.keywords.apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies.cast.apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies.crew.apply(lambda x:x.replace(" ",""))

movies.crew[0]
movies.head()
```

```python
# make column named tag

movies['crew']=movies['crew'].apply(lambda x:[x])
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head()
```

```python
# discarding the unnecessary columns

new_df=movies[['id','title_x','tags']]
```

```python
new_df
```

```python
# joining the elements to form string in tags columns

new_df['tags'] = new_df['tags'].str.join(' ') # method to condense into string
new_df.head()
new_df['tags'][0]
```

```python
# import nltk library(to be used for stemming)
import nltk
```

```python
# import stemPorter

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
# define stem function to do stemming in tags

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
# apply stem function on tags column

new_df.tags=new_df.tags.apply(stem)
```

```python
# import count-vectorizer

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
# define countVectorizer

vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
len(cv.get_feature_names_out())
```

```python
# import cosine_similarity for similarity evaluation

from sklearn.metrics.pairwise import cosine_similarity
```

```python
# defining the similarity matrix

similarity=cosine_similarity(vectors)
similarity[0].shape
```

```python
# define recommned function --  fetches top 5 movies based on cosine_similarity score

def recommend(movie):
    movie_index=new_df[new_df['title_x']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(i)
```

```python
# testing the recommend function

recommend('Her')
```
